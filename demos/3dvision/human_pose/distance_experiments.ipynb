```python
from scipy import misc as scpy
import skimage.feature as edge
import scipy.ndimage.morphology as scp_morph
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

matplotlib.rcParams['figure.figsize'] = (20.0, 20.0)
matplotlib.rcParams['figure.figsize'] = (5.0, 5.0)
%matplotlib inline
# %matplotlib notebook
```

```python
def compare(fn1, fn2):
    im1 = scpy.imread(fn1,True)/255.
    im2 = scpy.imread(fn2,True)/255.
    f, axarr = plt.subplots(5, 2)
    axarr[0,0].imshow(im1)
    axarr[0,1].imshow(im2)
    canny1 = edge.canny(im1)
    canny2 = edge.canny(im2)
    axarr[1, 0].imshow(canny1)
    axarr[1, 1].imshow(canny2)
    axarr[3, 0].imshow(canny1 | canny2)
    dist1 = scp_morph.distance_transform_edt(np.invert(canny1))
    dist2 = scp_morph.distance_transform_edt(np.invert(canny2))
    axarr[2, 0].imshow(dist1)
    axarr[2, 1].imshow(dist2)
    axarr[3, 1].imshow(dist1 + canny2*np.max(dist1))
    D = dist1 * canny2
    im = axarr[4,1].imshow(D)
    f.colorbar(im)
    axarr[4, 0].hist(D[canny2>0].flatten(), bins=100)
    
def compare_new_fig(fn1, fn2):
    im1 = scpy.imread(fn1,True)/255.
    im2 = scpy.imread(fn2,True)/255.
    im1c = mpimg.imread(fn1)
    im2c = mpimg.imread(fn2)
    plt.imshow(im1c)
    plt.figure()
    plt.imshow(im1)
    plt.figure()
    plt.imshow(im2c)
    canny1 = edge.canny(im1)
    canny2 = edge.canny(im2)
    plt.figure()
    plt.imshow(canny1)
    plt.figure()
    plt.imshow(canny2)
    plt.figure()
    plt.imshow(canny1 | canny2)
    dist1 = scp_morph.distance_transform_edt(np.invert(canny1))
    dist2 = scp_morph.distance_transform_edt(np.invert(canny2))
    plt.figure()
    plt.imshow(dist1)
    plt.figure()
    plt.imshow(dist2)
    plt.figure()
    plt.imshow(dist1 + canny2*np.max(dist1))
    D = dist1 * canny2 
    D2 = D - 4 * canny1 *canny2
    plt.figure()
    im = plt.imshow(D)
    plt.colorbar(im)
    plt.figure()
    im = plt.imshow(D2)
    plt.colorbar(im)
    plt.figure()
    im = plt.imshow(canny1*canny2)
    plt.colorbar(im)
    plt.figure()
    plt.hist(D[canny2>0], bins=100)
    plt.figure()
    plt.hist(D2[canny2>0], bins=100)
    print (D[canny2>0].mean(), D2[canny2>0].mean())
```

```python
compare_new_fig("examples/small/07.png",
    "samples/07.png/0000/sample_00222.png")
```
